## Rasters

```python
# Carregue o DEM a partir de um arquivo
with rasterio.open(dem_path) as dem:
    elevation = dem.read(1)  # Lê a primeira banda
    print("Metadata:", dem.meta)
    plt.imshow(elevation, cmap='viridis')
    plt.colorbar(label='Elevação (m)')
    plt.title('Modelo Digital de Elevação')
    plt.show()
```

```python
# Carregar o arquivo ASA
with rasterio.open(asa_path) as asa:
    asa_data = asa.read(1)  # Lê a primeira banda
    plt.imshow(asa_data, cmap='viridis')
    plt.colorbar(label='Amplitude do Sinal Analítico')
    plt.title('Amplitude do Sinal Analítico (ASA)')
    plt.show()
```

```python
# Exemplo para Potássio (K)
with rasterio.open(k_path) as k:
    k_data = k.read(1)
    plt.imshow(k_data, cmap='viridis')
    plt.colorbar(label='Potássio (K)')
    plt.title('Potássio (K)')
    plt.show()
```

```python
# Carregar o arquivo eU
with rasterio.open(eu_path) as eu:
    eu_data = eu.read(1)  # Lê a primeira banda
    plt.imshow(eu_data, cmap='viridis')
    plt.colorbar(label='Concentração de Urânio (eU)')
    plt.title('Concentração de Urânio (eU)')
    plt.show()
```

```python
# Carregar o arquivo eTh
with rasterio.open(eth_path) as eth:
    eth_data = eth.read(1)  # Lê a primeira banda
    plt.imshow(eth_data, cmap='viridis')
    plt.colorbar(label='Concentração de Tório (eTh)')
    plt.title('Concentração de Tório (eTh)')
    plt.show()
```
